```python
print(rows)
```

```python
df = pd.DataFrame.from_records(rows)
```

```python
df = pd.DataFrame.from_records(rows, columns=rows[0])
```

```python
df = df[df['XTB'] == 'tak']
```

```python
df = df.drop(columns=['Dodano dnia','mBank','XTB','BOSSA','Kurs','Waluta', 'MIN 1R', 'MAX 1R', 'NOW/MAX', 'MIN/MAX'])
```

```python
df = df[df['Replikacja'] == 'Fizyczna']
```

```python
df = df[df['Typ'] == 'Accumulating']
```

```python
df['Opłata roczna'] = df['Opłata roczna'].str.replace('%','')
```

```python
df['Opłata roczna'] = df['Opłata roczna'].str.replace(',','.')
```

```python
df['DY 2019'] = df['DY 2019'].str.replace('%','')
```

```python
df['DY 2019'] = df['DY 2019'].str.replace(',','.')
```

```python
df['DY 2019'] = df['DY 2019'].str.replace('','0')
```

```python
df['DY 2020'] = df['DY 2020'].str.replace('%','')
```

```python
df['DY 2020'] = df['DY 2020'].str.replace(',','.')
```

```python
df['DY 2020'] = df['DY 2020'].str.replace('','0')
```

```python
df['DY 2021'] = df['DY 2021'].str.replace('%','')
```

```python
df['DY 2021'] = df['DY 2021'].str.replace(',','.')
```

```python
df['DY 2021'] = df['DY 2021'].str.replace('','0')
```

```python
df['DY 2022'] = df['DY 2022'].str.replace('%','')
```

```python
df['DY 2022'] = df['DY 2022'].str.replace(',','.')
```

```python
df['DY 2022'] = df['DY 2022'].str.replace('','0')
```

```python
df['Aktywa\n(mln EUR)\n(05.23)'] = df['Aktywa\n(mln EUR)\n(05.23)'].str.replace('\xa0','')
```

```python
df = df.astype({'DY 2019':'float', 'DY 2020':'float', 'DY 2021':'float', 'DY 2022':'float','Aktywa\n(mln EUR)\n(05.23)':'float', 'Opłata roczna':'float'})
```
